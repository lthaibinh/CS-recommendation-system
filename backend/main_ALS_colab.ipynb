### ⚠️ Important Note: ALS Model Usage in `implicit` Library

**Correct Usage for ALS:**

The ALS implementation in `implicit` expects **user×item** matrices for both training and inference:


**Key Points:**
- **DO NOT transpose** the matrix for ALS (unlike some other libraries)
- Rows represent users, columns represent items
- For 21,202 users and 24,232 items: shape is **(21202, 24232)**
- The model treats the first dimension as users, second as items

Dữ liệu 2.5 triệu dòng của bạn là một log giao dịch. Mô hình ALS không đọc được log, nó cần một ma trận (User, Item, Rating). Chúng ta cần tạo ra cột "Rating" ngầm định.

Chúng ta sẽ không dùng Price làm rating. Thay vào đó, chúng ta sẽ đếm số lần mua (purchase_count). Đây là tín hiệu "độ tin cậy" (confidence) của chúng ta. Một người mua 10 lần thì "thích" sản phẩm đó hơn một người mua 1 lần.

Bước 2: Phân chia Dữ liệu (Train / Validation / Test)

```python
(train_data, validation_data, test_data) = data_agg.randomSplit([0.7, 0.15, 0.15], seed=42)
```

Bước 3: Khởi tạo Mô hình ALS

3.1 Defind model

```python
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RankingEvaluator
from itertools import product
import pyspark.sql.functions as F
from pyspark import StorageLevel
# BjanSparkSession is already available as 'spark'

# --- Chuẩn bị true_items validation (một lần)
true_items_val = validation_data.groupBy("Account_Id") \
    .agg(F.collect_list("ProductId").alias("true_items"))

val_users = validation_data.select("Account_Id").distinct()

# --- CRITICAL: Cache các DataFrame để tránh StackOverflowError ---
# Cache để tránh recompute và giảm lineage depth
train_data.cache()
val_users.cache()
true_items_val.cache()

# Trigger materialization
print(f"Training data count: {train_data.count()}")
print(f"Validation users count: {val_users.count()}")
print(f"True items count: {true_items_val.count()}")

# --- Evaluator
evaluator = RankingEvaluator(
    metricName="ndcgAtK",
    k=10,
    labelCol="true_items",
    predictionCol="pred_items"
)

# --- Grid tham số
ranks = [10] # [10, 20, 30, 40]
regParams = [ 1] #[0.01, 0.1, 1]
alphas =  [40] # [1, 10, 20, 40]
maxIters = [10] # [10, 15]

param_grid = [
    {"rank": r, "regParam": rp, "alpha": a, "maxIter": mi}
    for r, rp, a, mi in product(ranks, regParams, alphas, maxIters)
]

print(f"Total number of parameter combinations: {len(param_grid)}")

# --- Định nghĩa hàm train với cleanup để tránh StackOverflowError ---
def train_and_eval(params, train_df, val_users_df, true_items_df, eval_instance):
    """
    Trains one ALS model with the given params and evaluates it.
    This function runs on the DRIVER.
    """
    als = ALS(
        userCol="Account_Id",
        itemCol="ProductId",
        ratingCol="purchase_count",
        implicitPrefs=True,
        coldStartStrategy="drop",
        **params
    )

    print(f"--- Training with params: {params} ---")

    # als.fit() LÀ MỘT SPARK JOB, nó sẽ tự động chạy song song
    model = als.fit(train_df)

    # Các lệnh recommend và join cũng là Spark jobs
    pred_items_val = model.recommendForUserSubset(val_users_df, 10) \
        .selectExpr("Account_Id", "transform(recommendations, x -> x.ProductId) as pred_items")

    eval_data = pred_items_val.join(true_items_df, "Account_Id") \
        .withColumn("true_items", F.expr("transform(true_items, x -> double(x))")) \
        .withColumn("pred_items", F.expr("transform(pred_items, x -> double(x))"))

    # Cache eval_data và trigger evaluation ngay để break lineage
    eval_data.cache()
    ndcg = eval_instance.evaluate(eval_data)

    # CRITICAL: Unpersist để free memory và avoid lineage accumulation
    eval_data.unpersist()

    # Clean up model (important to avoid memory issues)
    del model

    print(f"Params: {params} -> NDCG@10: {ndcg}")
    return (params, ndcg)

# --- Chạy tuning tuần tự (Sequential) trên Driver ---
results = []
for i, params in enumerate(param_grid, 1):
    print(f"\n{'='*60}")
    print(f"Progress: {i}/{len(param_grid)}")
    print(f"{'='*60}")

    # Truyền các DataFrame đã cached vào hàm.
    result = train_and_eval(
        params,
        train_data,       # DataFrame đã cached
        val_users,        # DataFrame đã cached
        true_items_val,   # DataFrame đã cached
        evaluator         # Evaluator object
    )
    results.append(result)

# --- Clean up cached data ---
train_data.unpersist()
val_users.unpersist()
true_items_val.unpersist()

# --- Chọn model tốt nhất
best_params, best_ndcg = max(results, key=lambda x: x[1])
print("\n" + "="*60)
print(f"🏆 Best params: {best_params}")
print(f"🏆 Best NDCG@10: {best_ndcg}")
print("="*60)
```
